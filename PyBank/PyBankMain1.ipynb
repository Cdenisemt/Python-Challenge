```python
#imports
import os
import csv
from pathlib import Path

#file paths 
budget_data_csv=os.path.join('Resources', 'budget_data.csv')

output_path = os.path.join("output", "financial_analsysis.txt")

csv_path = Path("budget_data.csv")

with open (budget_data_csv) as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')

#initialize variables 
total_months = 0
net_total = 0
previous_profit= 0
profit = 0 
profit_changes = []
dates = []
date = []

#read csv file 
with open(budget_data_csv) as csvfile:
    csvreader = csv.reader(csvfile)
    header = next(csvreader)

    for row in csvreader:
        date = row[0]
        profit = int(row[1])
   
#calculate total months and net total

        total_months += 1
        net_total += profit

#calculate profit change

        if previous_profit is not None:
            profit_change = profit - previous_profit
            profit_changes.append(profit_change)
            dates.append(date)

        previous_profit = profit

#calculate average change

if profit_changes:
    average_change = sum(profit_changes) / len(profit_changes)
else:
    average_change = 0

#find greatest increase and decrease

if profit_changes:
    greatest_increase = max(profit_changes)
    greatest_decrease = min(profit_changes)
    increase_date = dates[profit_changes.index(greatest_increase)]
    decrease_date = dates[profit_changes.index(greatest_decrease)]
else:
    greatest_increase = greatest_decrease = 0
    increase_date = decrease_date = "N/A"

#prepare the analysis results
analysis =f"""
Financial analysis
---------------------------------
Total Months:{total_months}
Total: ${net_total}
Average Change: ${average_change:.2f}
Greatest Increase in Profits: {increase_date} (${greatest_increase})
Greatest Decrease in Profits: {decrease_date} (${greatest_decrease})
"""

#print analysis 
print(analysis)

#export analysis

output_file = os.path.join("analysis.csv")

with open(output_path, "w") as file:
   file.write(analysis)

print(f"\nResults have been exported to {output_path}")






```
